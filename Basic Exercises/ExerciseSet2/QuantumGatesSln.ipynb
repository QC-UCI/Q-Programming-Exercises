```python
from qiskit import QuantumCircuit, transpile, Aer, IBMQ, assemble, execute, QuantumRegister, ClassicalRegister
from qiskit.tools.jupyter import *
from qiskit.visualization import *
from qiskit.providers.aer import QasmSimulator
import qiskit.quantum_info as qi

from qiskit_ibm_provider import IBMProvider
import qiskit_ibm_provider

import seaborn
import numpy as np

provider = IBMProvider()
```

```python
### UTILITY ###

pi = np.pi

def aer_simulate_circuit(qc):
    backend = Aer.get_backend('qasm_simulator')
    job = execute(qc, backend, shots = 1000)
    result = job.result()         
    return result.get_counts()

### NOTES ###
# Note that Qiskit is Little-Endian, which means the least significant digit is on the left. 
# In, |001>, the first qubit is the rightmost (w/ value = 1)
# |3rd, 2nd, 1st>
```

### Superposition

```python
qc = QuantumCircuit(4)
qc.i(0)
qc.ry(pi/6, 1) # (theta, qubit)
qc.ry(pi/3, 2) # (theta, qubit)
qc.h(3)

qc.draw()
```

```python
plot_bloch_multivector(qc, title="Rotating Statevector along y-axis")
```

```python
qc.x([i for i in range(4)])
qc.draw()
```

```python
plot_bloch_multivector(qc, title="Statevector reflecting upon x-axis")
```

```python
qc.rz(pi/2, [i for i in range(4)])
qc.draw()
```

```python
plot_bloch_multivector(qc, title="Statevector rotating π/2 rad around z-axis")
```

### Entanglement

```python
qce_demo = QuantumCircuit(2)
qce_demo.h(0)
qce_demo.cx(0,1)
qce_demo.measure_all()
qce_demo.draw()
```

```python
# What should we expect?
# Initial: |00>
#    H(0): 1/√2|00> + 1/√2|01>
# CX(0,1): 1/√2|00> + 1/√2|11>

plot_histogram(aer_simulate_circuit(qce_demo)) 
```

```python
# [1] #

# Initial: |0000>
#    H(0): 1/√2(|0000> + |0001>)
# CX(0,1): 1/√2(|0000> + |0011>)
# CX(0,2): 1/√2(|0000> + |0111>)
# CX(0,3): 1/√2(|0000> + |1111>)
```

```python
# [2] #

# Initial: |0000>
#    X(1): |0010>
#    H(3): 1/√2(|1010> + |0010>)
#    X(3): 1/√2(|0010> + |1010>)
# CX(3,2): 1/√2(|0010> + |0110>)
#    X(3): 1/√2(|1010> + |0110>)
#    Z(2): 1/√2(|1010> - |0110>)
```

```python
qc2 = QuantumCircuit(4)
qc2.x(1)
qc2.h(3)
# Zero-controlled CX
qc2.x(3)
qc2.cx(3,2)
qc2.x(3)
#
qc2.z(2)
qc2.measure_all()

plot_histogram(aer_simulate_circuit(qc2)) 
```

```python
# [3] #

# Initial: |000>
#    H(0): 1/√2|000> + 1/√2|001>
# CH(0,1): 1/√2|000> + 1/√2(1/√2|001> + 1/√2|011>) 
#        = 1/√2|000> + 1/2|001> + 1/2|011> 
# CX(1,2): 1/√2|000> + 1/2|001> + 1/2|111> 
#    Z(2): 1/√2|000> + 1/2|001> - 1/2|111> 
```

```python
qc3 = QuantumCircuit(3)
qc3.h(0)
qc3.cry(pi/2, 0,1) #Controlled Hadamard
qc3.cx(1,2)
qc3.z(2)
qc3.measure_all()

plot_histogram(aer_simulate_circuit(qc3)) 
```
